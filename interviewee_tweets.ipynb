```python
import os
import json
from json_to_df_for_interviewee_tweets import json_to_df
from dateutil.parser import parse

begin_date = parse("2016-04-15")
end_date = parse("2016-06-22")

interviewees = ["euromove"]

interviewee_tweets = []

for filename in os.listdir("."):
    if os.path.isfile(filename) and filename.endswith(".json"):
        parsed_date = parse(filename.replace(".json", ""))
        if parsed_date >= begin_date and parsed_date <= end_date:
            print("processing: " + filename)
            with open(filename, "r") as f:
                for line in f:
                    try:
                        tweet = json.loads(line)
                        username = tweet["user"]["screen_name"]
                        if username.lower() in interviewees:
                            interviewee_tweets.append(line)
                            print(username)
                    except (ValueError, KeyError):
                        pass

with open("euromove_tweets.json", "w") as outfile:
    outfile.write("".join(interviewee_tweets))
```
