### Group Values

What is the average RT and like count for every hour of the day and every day of the week? What is the retweeted frequency for every hour of the day and day of the week? Only report retweet count because of the missing values problem for like count.

```python
grouped_hour = df.groupby('hour')
grouped_hour.size()
```

```python
grouped_hour[['rt_count', 'like_count']].agg(np.mean)
```

```python
fig = plt.figure(figsize=(16,8))
sns.boxplot(x='hour', y='rt_count', data=df)
plt.title('Retweet Count by Hour of Day')
plt.ylabel('Retweet Count')
plt.xlabel('Hour of Day')
plt.ylim([0,6]);
plt.savefig('Retweet Count by Hour of Day.jpg', bbox_inches='tight')
```

```python
grouped_hour['rt_count'].agg(np.mean).plot(kind='bar', figsize=(16,8), legend=False)
plt.title('Mean Retweet Count for Hour of Day')
plt.ylabel('Mean Retweet Count')
plt.xlabel('Hour of Day');
#plt.savefig('Mean Retweet Count for Hour of Day.jpg', bbox_inches='tight')
```

```python
pd.crosstab(df.hour, df.rt_count_binary).plot(kind='bar', figsize=(16,8))
plt.title('Retweeted Frequency for Hour of Day')
plt.xticks(rotation='horizontal')
plt.xlabel('Hour of Day');
plt.savefig('Retweeted Frequency for Hour of Day', bbox_inches='tight')
```

```python
grouped_hour_binned = df.groupby('hour_binned')
grouped_hour_binned.size()
```

```python
grouped_hour_binned[['rt_count', 'like_count']].agg(np.mean)
```

```python
fig = plt.figure(figsize=(10,6))
sns.boxplot(x='hour_binned', y='rt_count', data=df)
plt.title('Retweet Count by Hour (Binned)')
plt.ylabel('Retweet Count')
plt.xlabel('Hour (Binned)')
plt.ylim([0,6]);
plt.savefig('Retweet Count by Hour (Binned).jpg', bbox_inches='tight')
```

```python
grouped_day = df.groupby('day')
grouped_day.size()
```

```python
grouped_day[['rt_count', 'like_count']].agg(np.mean)
```

```python
fig = plt.figure(figsize=(12,8))
sns.boxplot(x='day', y='rt_count', data=df)
plt.title('Retweet Count by Day of Week')
plt.ylabel('Retweet Count')
plt.xlabel('Day of Week')
plt.ylim([0,6]);
plt.savefig('Retweet Count by Day of Week.jpg', bbox_inches='tight')
```
